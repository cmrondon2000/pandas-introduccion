##**¿Cuál es el porcentaje de cada tipo de propiedad en nuestra base de datos?**

```python
datos.Tipo.unique()
```

```python
inmuebles_comerciales = ['Conjunto Comercial/Sala', 'Edificio Completo',
       'Tienda/Salón', 'Casa Comercial', 'Cochera/Estacionamiento',
       'Galpón/Depósito/Almacén', 'Tienda en Centro Comercial', 'Rancho',
       'Posada/Chalé', 'Hotel', 'Loteo/Condominio',
       'Industria']
```

```python
datos.query("@inmuebles_comerciales in Tipo")
```

```python
datos.query("@inmuebles_comerciales not in Tipo")
```

```python
df = datos.query("@inmuebles_comerciales not in Tipo")
df_tipo_precio = df.groupby("Tipo")[["Valor"]].mean().sort_values("Valor", ascending=False)
df_tipo_precio.plot(kind="bar", figsize=(12, 8), color = "purple")
```

```python
df.Tipo.unique()
```

```python
df.Tipo.value_counts()
```

```python
df.Tipo.value_counts(normalize=True)
```

```python
df.query("Tipo == 'Departamento'")
```

```python
df = df.query("Tipo == 'Departamento'")
df.head(5)
```

```python
porcentaje_tipo = df.Tipo.value_counts(normalize=True).to_frame()
porcentaje_tipo.plot(kind="bar", figsize=(12, 8), color = "green", xlabel="Tipo", ylabel="Porcentaje")
```

```python
df = df.query("Tipo == 'Departamento'")
df.head(5)
```

```python
# Guardando el DataFrame en una variable
df_ejemplo = df['Tipo'].value_counts(normalize=True).to_frame().sort_values('Tipo')

# Cambiando el nombre de la columna "Tipo" a "Porcentajes"
df_ejemplo.rename(columns={'Tipo': 'Porcentajes'}, inplace=True)

# Visualizando el DataFrame
df_ejemplo
```
