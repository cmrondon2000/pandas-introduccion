#**¿Cuáles son los valores promedio de alquiler por tipo de propiedad?**

```python
datos.head()
```

```python
datos["Valor"].mean()
```

```python
datos.groupby("Tipo").mean(numeric_only=True)
```

```python
datos.groupby("Tipo")["Valor"].mean()
```

```python
datos.groupby("Tipo")[["Valor"]].mean().sort_values("Valor", ascending=False)
```

```python
df_tipo_precio = datos.groupby("Tipo")[["Valor"]].mean().sort_values("Valor", ascending=False)
```

```python
df_tipo_precio.plot(kind="bar", figsize=(12, 8), color = "purple")
```
